# Serbyn calculation in 1D

$$Q(0,0)=-e^2 \frac{1}
    {2\beta} \sum_{\epsilon_n}\frac{1}{L}\sum_k Tr\left[ v^2 G(k,i\epsilon_n)G(k,i\epsilon_n)\right]$$

$$G(k, i\epsilon_n)=\left[\left(i\epsilon_n-\int d\omega \frac{\gamma}{\omega+i\epsilon_n}\right)\tau_0\sigma_0-\left(2w_0cos(k)-\mu\right)\tau_z\sigma_0+\Delta\tau_y\sigma_y\right]^{-1}$$
$$\epsilon_n-\epsilon^{(1)}=-\Gamma$$

$$\epsilon_n=\frac{2\pi}{\beta}\left(n+\frac{1}{2}\right)$$

$$\Delta=\Delta^{(1)}$$

```python
import numpy as np
import matplotlib.pyplot as plt
from Serbyn_1D import get_Q, get_Q_k
```

```python
beta = 100
N = 100
epsilon_n = 2*np.pi/beta*(np.arange(N)+1/2)
L = 100
k = 2*np.pi/L*np.arange(0, L)
w_0 = 1
Gamma = 0.01
Delta = np.linspace(-1, 1, 10)
Delta = np.concatenate((Delta[Delta<0], [0], Delta[Delta>0]))
mu = 0

Q_vs_Delta = [get_Q(k, epsilon_n, w_0, Gamma, Delta, mu, beta) for Delta in Delta]

fig, ax  = plt.subplots()
ax.plot(Delta, Q_vs_Delta)
ax.set_xlabel(r"$\Delta$");
ax.set_ylabel(r"$Q(\Delta)$");
```
