# 1D tight-binding

## Hamiltonian
$$H_S(t) = \sum_k \varepsilon_0 c^{\dagger}_k c_k + \sum_k w_0 \left(e^{i k} e^{i \phi t}+e^{-i k} e^{-i \phi t}\right)c^{\dagger}_k c_k$$

## DC-conductivity

$$\sigma=
 -4 w_0^2 \sum_k   \sin^2 (k) \int \frac{d \omega}{2\pi} \partial_\omega  f(\omega) \left[  \rho^f_k(0,\omega)\right]^2$$
 
 At zero temperature $\partial_\omega  f(\omega)=\delta(\omega)$, then it becomes:
 $$\sigma=
 -\frac{2}{\pi} w_0^2 \sum_{k=-\pi}^{\pi}   \sin^2 (k) \left[  \rho^f_k(0,0)\right]^2$$
 
 with
 
 $$ \rho^f_k(\theta,\omega) = G_k^f(\theta,\omega) \Gamma [G_k^f(\theta,\omega)]^\dagger,$$
 
$$G^f_{k}(t,\omega)=\left[\omega - \varepsilon_0-w_{k}( t) +i {\Gamma} \right]^{-1}$$

and

$$w_k(t)=2 w_0 \cos(k+\phi t)$$

```python
import numpy as np
import matplotlib.pyplot as plt
from _1D_tight_binding import get_sigma, Rho_k


k = np.linspace(-np.pi, np.pi, 1000)
E_0 = 0
w_0 = 1
Gamma = 0.1

sigma = get_sigma(k, 0, E_0, w_0, Gamma)
rho_k = [np.real(Rho_k(k, 0, E_0, w_0, Gamma)) for k in k]
fig, ax  = plt.subplots()
ax.plot(k, rho_k)
ax.set_xlabel(r"$k$");
ax.set_ylabel(r"$\rho^f_k(\omega=0)$");
```
